### renaming the column we need

### Comverting the type of the TWO new columns (publish_time / publish_date)

```python
df['publish_date'] = pd.to_datetime(df['publish_date'])
df
```

```python
df.dtypes
```

```python
# convert it again
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
```

```python
df['day_name'] = df['trending_date'].dt.day_name()
df
```

### Calculate Time to Trend - Create a new column that shows how many days it took for a video to start trending after it was published.

```python
df['publish_date'] = pd.to_datetime(df['publish_date'])
df['trending_date'] = pd.to_datetime(df['trending_date'])

# Calculate time difference between trending and publishing
df['days_to_trend'] = (df['trending_date'] - df['publish_date']).dt.days
df
```

### Convert to hour

```python
df['publish_day'] = df['publish_date'].dt.day_name()
df['publish_hour'] = pd.to_datetime(df['publish_time'], format='%H:%M:%S').dt.hour
```

```python
df
```

### first step: calculating average days to trend by publishing day (how long on average it takes for videos to start trending based on the day it has been published)

```python
avg_days_by_day =df.groupby('publish_day')['days_to_trend'].mean().sort_values().round()
print(avg_days_by_day)
```

```python
avg_days_by_day.plot(kind='bar', color='purple', title='The Impact of Publish Day on Video Trending Time')
plt.xlabel('Publish Day')
plt.ylabel('Number of days to be Trend')
```

### Second step: Average days to trend by hour (how many days it takes for a video to trend based on the hour it was published

```python
avg_trend_by_hour = df.groupby('publish_hour')['days_to_trend'].mean().sort_index().round()
print(avg_trend_by_hour)
```

```python
avg_trend_by_hour.plot(kind='bar', color='purple', title='The Impact of Publish Time on Video Trending Time')
plt.xlabel('Publish Hour')
plt.ylabel('Number of days to be Trend')
```

## Q5: Which channels trend the most — and what do they have in common?

### step1: Identify the top trending channels

```python
top_channels = df['channel_title'].value_counts().head(10)
print(top_channels)
```

```python
top_channels.plot(kind='bar', color='purple', title='Top 10 Most Trending Channels in US YouTube')
plt.xlabel('Channel Title')
plt.ylabel('Number of Published Videos')
```

### Step2: What channels have in common 

#### 1. Category

```python
top_df = df[df['channel_title'].isin(top_channels.index)]

common_categories = top_df.groupby('channel_title')['category_name'].agg(lambda x: x.mode()[0])
common_categories_df = pd.DataFrame(common_categories)
common_categories_df = common_categories_df.reset_index()
#common_categories_df['common category']=common_categories_df['category_name']
common_df=pd.DataFrame(common_categories_df['category_name'].value_counts())
common_df
```

```python
common_df.plot(kind='bar', color='purple', title='Common Categories in Trending Channels')

plt.xlabel('Category Name')
plt.ylabel('Number of Channels')
```

####  2. publishing time patterns (Do they upload at similar hours?)

```python
upload_hours = top_df.groupby('channel_title')['publish_hour'].mean().sort_values().round()
print(upload_hours)
```

```python
#upload_hours.reset_index()
sns.barplot(upload_hours.reset_index(), x='publish_hour', y='channel_title', hue='publish_hour')
plt.xlabel('Upload Hour')
plt.ylabel('Channel Title')
plt.title('Common Upload Hours Among Trending Channels')
```

#### 3. common publishing day

```python
common_days = top_df.groupby('channel_title')['publish_day'].agg(lambda x: x.mode()[0])
common_days_df = pd.DataFrame(common_days).reset_index()
common_df = pd.DataFrame(common_days_df['publish_day'].value_counts())

common_df
```

```python
top_df = df[df['channel_title'].isin(top_channels.index)]

common_days = top_df.groupby('channel_title')['publish_day'].agg(lambda x: x.mode()[0])
common_days_df = pd.DataFrame(common_days)

print(common_days_df)
```

## Q:6	What are the most common tags or keywords among trending videos?

###  1. Combine all tags into one big list

```python
all_tags = [tag.lower() for tags in df['tags'].dropna() for tag in tags.split('|')]
```

```python
all_tags
```

### 2. Count how often each tag appears

```python
from collections import Counter
import pandas as pd

tag_counts = Counter(all_tags)
top_tags = pd.Series(tag_counts).sort_values(ascending=False).head(20)
```

```python
print(top_tags)
```

```python
# removing the null values
filtered_tags = [tag for tag in all_tags if tag.strip() not in ['[none]', '', 'null']]
tag_counts = Counter(filtered_tags)
top_tags = pd.Series(tag_counts).sort_values(ascending=False).head(20)
```

```python
top_tags
```

```python
top_tags.plot(kind='bar', color='purple', title='Top 20 Most Common Tags in Trending Videos')

plt.xlabel('tag')
plt.ylabel('Frequency Count of Each Tag')
```
